```python
import pandas as pd
import gzip
import shutil
```

```python
# Unzip and convert into text file
path = 'dailies/2021-04-27/'
filename = '2021-04-27_clean-dataset.tsv.gz'

with gzip.open(path+filename, 'rb') as f_in:
    with open('clean-dataset.tsv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
        
df_clean=pd.read_csv('clean-dataset.tsv', delimiter='\t')

with open('clean_txt_test.txt', 'w') as f:
    for tweet in df_clean.tweet_id:
        f.write(str(tweet)+'\n')
```

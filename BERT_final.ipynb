```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
# names = os.path.join("content/drive/MyDrive/COVID_country_differences/'BERT Output'/", "*.csv")
# files = glob.glob(names)

all_files = os.listdir()
csv_files = list(filter(lambda f: f.endswith('.csv'), all_files))

files = [x.split('.')[0] for x in csv_files]
files
```

Threshold 0.5 and 0.16

```python
col_list = ['Week', 'total_tweets','count_sadness', 'avg_sadness', 'count_joy', 'avg_joy', 'count_love', 'avg_love', 'count_anger', 'avg_anger', 'count_fear',\
            'avg_fear', 'count_surprise', 'avg_surprise']
df_tmp = pd.DataFrame(columns=col_list)

for filename in files:
  # filename = 'week_1'
  df = pd.read_csv(filename+'.csv')
  row_list = []
  row_list.append(filename)
  row_list.append(df.shape[0])

  for cols in ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']:
    df['is_'+cols] = [1 if x>0.5 else 0 for x in df[cols]]
    row_list.append(np.sum(df['is_'+cols]))
    row_list.append(np.mean(df[cols]))

  df_tmp.loc[len(df_tmp.index)] = row_list
```

```python
for cols in ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']:
  df_tmp['prop_'+cols] = df_tmp['count_'+cols]/df_tmp['total_tweets']
```

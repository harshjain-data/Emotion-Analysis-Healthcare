```python
df = pd.read_csv('12WEEKS.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df['clean_text'] = df['clean_text'].str.replace('[^\w\s]','')
```

```python
#Removal of stop words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
```

```python
#stop.append("covid vaccine")
new = ("covid19 vaccine", "covid19", "vaccine", "coronavirus", "vaccination", "covid", "corona"
      , "virus", "covid vaccine", "covidvaccine", "covid19vaccine")
stop.extend(new)
#stop.append("covid19 vaccine")
```

```python
#stop.pop()
stop[-10:]
```

```python
df['Tokens']  = df['clean_text'] .apply(lambda x: " ".join(x for x in x.split()  if x not in stop))
df.head()
```

```python
df = df[df['clean_text'].notnull()]
```

```python
df.count()
```

```python
NUM_SAMPLES = 50000 # The number of sample to use 
sample_df = df.sample(NUM_SAMPLES, replace=False).reset_index(drop=True)
```

```python
sample_df.shape
```

```python
sample_df.head()
```

```python
cv = CountVectorizer(max_df=0.75, min_df=2)
dtm = cv.fit_transform(sample_df['Tokens'])
```

```python
dtm
```

```python
# We can also get all those words using the get_feature_names() function
feature_names = cv.get_feature_names()
len(feature_names) # show the total number of distinct words
```

```python
temp_file = pd.DataFrame(feature_names)
```

```python
len(temp_file[0].unique())
```

```python
feature_names
```

```python
NUM_TOPICS = 20
LDA_model = LatentDirichletAllocation(n_components=NUM_TOPICS, max_iter=30, random_state=42)
```

```python
LDA_model.fit(dtm)
```

```python
len(feature_names)
```

```python
import random 
for index in range(15):
    random_word_ID = random.randint(0, 45)
    print(cv.get_feature_names()[random_word_ID])
```

```python
len(LDA_model.components_[0])
```

```python
# Pick a single topic 
a_topic = LDA_model.components_[0]

# Get the indices that would sort this array
a_topic.argsort()
```

```python
top_20_words_indices = a_topic.argsort()[-20:]

for i in top_20_words_indices:
    print(cv.get_feature_names()[i])
```

```python
for i, topic in enumerate(LDA_model.components_):
    print("THE TOP {} WORDS FOR TOPIC #{}".format(20, i))
    print([cv.get_feature_names()[index] for index in topic.argsort()[-20:]])
    print("\n")
```

```python
final_topics = LDA_model.transform(dtm)
final_topics.shape
```

```python
final_topics[0]
```

```python
final_topics[0].argmax()
```

```python
sample_df["Topic N°"] = final_topics.argmax(axis=1)
```
